```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

# 주피터 환경용 스파크 세션 생성
spark = SparkSession.builder \
    .appName("Animal-Data-Profiling") \
    .getOrCreate()

# Raw CSV 데이터 로드 (인프라 깨짐을 눈으로 확인하기 위해 inferSchema=True로 먼저 테스트)
raw_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("hdfs:///user/maria_dev/animal_project/raw/*.csv")

print(f"✨ 데이터 로드 완료! 전체 행 개수: {raw_df.count()}개 | 컬럼 개수: {len(raw_df.columns)}개")
```
